## Diabetes Prediction

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
#independent and dependent Feature
X=df_diabetes
y=dataset['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)

```

## Hyperparameter Tuning

```python
param={'criterion':['friedman_mse', 'absolute_error', 'squared_error', 'poisson'],
       'splitter':['best', 'random'],
       'max_depth':[1,2,3,4,5,6],
       'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# predictions
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("MSE:",mse)
print("MAE:",mae)
print("Score:",score)
```

```python
selectmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='sqrt',splitter='random')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
## Visualize the Desicion Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(16,10))
tree.plot_tree(selectmodel,filled=True)
```
