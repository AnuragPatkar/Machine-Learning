# Algerian Forest Fires Dataset

## üìå Data Set Information

The dataset includes **244 instances** from two regions of Algeria:
- **Bejaia Region** (northeast) ‚Äî 122 instances  
- **Sidi Bel-Abbes Region** (northwest) ‚Äî 122 instances  

üóìÔ∏è Period: **June 2012 to September 2012**  
üî• Classified into:
- **Fire:** 138 instances  
- **Not Fire:** 106 instances  

---

## üßæ Attribute Information

<br>

<font size="4"><b>Date:</b></font> (DD/MM/YYYY) ‚Äî Day, month ('June' to 'September'), year (2012)  
<font size="4"><b>Temp:</b></font> Temperature at noon (¬∞C): *22 to 42*  
<font size="4"><b>RH:</b></font> Relative Humidity (%): *21 to 90*  
<font size="4"><b>Ws:</b></font> Wind speed (km/h): *6 to 29*  
<font size="4"><b>Rain:</b></font> Rainfall (mm): *0 to 16.8*  

---

## üî• FWI Components

<font size="4"><b>FFMC:</b></font> Fine Fuel Moisture Code ‚Äî *28.6 to 92.5*  
<font size="4"><b>DMC:</b></font> Duff Moisture Code ‚Äî *1.1 to 65.9*  
<font size="4"><b>DC:</b></font> Drought Code ‚Äî *7 to 220.4*  
<font size="4"><b>ISI:</b></font> Initial Spread Index ‚Äî *0 to 18.5*  
<font size="4"><b>BUI:</b></font> Buildup Index ‚Äî *1.1 to 68*  
<font size="4"><b>FWI:</b></font> Fire Weather Index ‚Äî *0 to 31.1*  

---

## üéØ Output Class

- `Fire`
- `Not Fire`


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
#missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
##Removing the Null Value
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#Fix Spaces in Coloum name
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## Changes Required Coloums as integer Data Type


```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## Changing the Other Coloums to Float Data Type

```python
objects=[features for features in df.columns if df[features].dtype=='O' ]
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
